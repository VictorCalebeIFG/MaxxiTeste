```python
nome = ["CarolinÃ£ da Silva",
        "gabrielly rezenDe",
        "Bernardo Dias",
        "helena gonÃ§alves",
        "Renan Caldeira",
        "Lucca Lopes",
        "AndrÃ© Ribeiro",
        "JoÃ£o Marcelo",
        "Joana da Luz",
        "Vicente Moreira",
        "Camila d' Rocha",
        "Elisa Monteiro",
        "Daniela Dias",
        "danIel Gomes",
        "Heitor Nogueira"]

e_mail = [
    "carolina.sil@gmail.com",
    "gabi_rezende@google.com",
    "bernardodias@aws.com",
    "helena.gonÃ§alves.gmail.com",
    "renan_calDEIRA@google.com",
    "lucca.com.br",
    "andrÃ©_ribeiro.outlook.com.br",
    "joÃ£o_ma@outlook.br",
    "joluz@google.com.br",
    "vicente.moreira@maxxi.com",
    "milarocha@gmail.com",
    "Elisa_mo23@yahoo.com.bre",
    "danidias@maxxi.com",
    "danieldias@aws.com",
    "nogueira_heitor@gmail.com"
]


idade =[
        '35',
        '43',
        '70',
        '23',
        '61',
        '46',
        '39',
        '32',
        '22',
        '19',
        '51',
        '30',
        '43',
        '20',
        '64']
```

ð¡**Tradando os nomes:**

```python
import re
import unicodedata

# Decompondo Caractereres especiais em caracteres de combinaÃ§Ã£o.
nome_normalizado = [unicodedata.normalize('NFD',n) for n in nome]

# Removendo os Caracteres de CombinaÃ§Ã£o.
nome_tratado = [re.sub(r'[\u0300-\u036f]', '', nn).casefold() for nn in nome_normalizado]

# Mantendo Somente os Caracteres Alfa numÃ©ricos.
nome_tratado = [re.sub('[^A-Za-z0-9 ]+', '', nt) for nt in nome_tratado]
```

ð¡**Tradando os e-mails:**

```python
# Decompondo Caractereres especiais em caracteres de combinaÃ§Ã£o.
email_normalizado = [unicodedata.normalize('NFD',n) for n in e_mail]

# Removendo os Caracteres de CombinaÃ§Ã£o.
email_tratado = [re.sub(r'[\u0300-\u036f]', '', nn).casefold() for nn in email_normalizado]

# Mantendo Somente os Caracteres Alfa numÃ©ricos.
email_tratado = [re.sub('[^A-Za-z0-9 @._]+', '', nt) for nt in email_tratado]
```

ð¡**Coluna de ValidaÃ§Ã£o**

```python
# FunÃ§Ã£o que vai verificar se o e-mail Ã© valido ou nÃ£o.
def validar(email):
    if (email.endswith('.com') or  email.endswith('.com.br')) and '@' in email:
        return 1
    else: return 0

# Mapeando todos os e-mails usando a funÃ§Ã£o validar.
is_valid = list(map(validar,email_tratado))
```

```python
import pandas as pd

# Criando um dataframe com os dados tratados.
dados_tratados = pd.DataFrame({'nome':nome_tratado,
                               'idade':idade,
                               'email':email_tratado,
                               'email_valid':is_valid})

dados_tratados.head()
```
